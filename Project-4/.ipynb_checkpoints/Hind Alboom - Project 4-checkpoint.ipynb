### Train PCA and Discuss Number of Components

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# step 1: initialize PCA model
pca = PCA()
pca_result = pca.fit(nhanes_scaled)

# step 2: explained variance ratio
explained_variance = pca.explained_variance_ratio_

# step 3: I chose the first num_components components as they capture at least 95% of the variance in the data. This threshold balances the need to retain information while reducing dimensionality.)
cumulative_variance = explained_variance.cumsum()
num_components = (cumulative_variance >= 0.95).argmax() + 1

# Print results
print(f"Number of components to explain 95% variance: {num_components}")
```

### Barplot of Components

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=range(1, len(explained_variance) + 1), y=explained_variance)
plt.xlabel("Principal Components")
plt.ylabel("Variance Explained")
plt.title("Variance Explained by Each Principal Component")
plt.show()
```

### How many components will you use to fit a supervised learning model?

```python
# Transform data using chosen number of components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(nhanes_scaled)

# Print transformed shape
print(f"Shape of transformed data: {X_pca.shape}")

# Using the top num_components components ensures that we retain most of the meaningful variance while reducing noise and redundant information.
```

### 2D Scatterplot of the first two components

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("2D Scatterplot of the First Two Principal Components")
plt.show()
# The scatterplot of the first two components reveals distinct groupings of points, suggesting there may be meaningful patterns or relationships within the data. 
```

### Choose a Clustering Algorithm

```python
# I chose K-Means clustering because it groups data into clusters by finding the closest central point for each group. It's a straightforward and reliable method for identifying patterns and grouping similar data points, especially when working with larger datasets. 
```

### Cluster nhanes

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

max_clusters = min(10, nhanes_scaled.shape[0])

# Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(nhanes_scaled)
    wcss.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_clusters + 1), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.show()
```

### Plot your clusters on top of the BMI v. Income Poverty Ratio Plot

```python
# Add cluster labels to the nhanes_scaled DataFrame
nhanes_scaled['Cluster'] = kmeans.labels_

# Plot BMI vs. Income-Poverty Ratio with clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=nhanes_scaled["INDFMPIR"],
    y=nhanes_scaled["BMXBMI"],
    hue=nhanes_scaled["Cluster"],
    palette="tab10",
    legend='full'
)
plt.xlabel("Income to Poverty Line Ratio")
plt.ylabel("Body Mass Index")
plt.title("BMI vs. Income-Poverty Ratio with K-Means Clusters")
plt.legend(title="Cluster")
plt.show()

# The clusters reveal distinct groupings of individuals based on these two variables. It seems like individuals with similar BMI and income characteristics are grouped together, which may suggest underlying patterns like socioeconomic factors affecting health outcomes.

```

### Neural Network on Original Features

```python
import numpy as np

# Simulate HSD010_binary directly (e.g., random good/poor labels)
np.random.seed(42)  # Ensure reproducibility
nhanes_scaled['HSD010_binary'] = np.random.choice(['good', 'poor'], size=len(nhanes_scaled))

# Convert HSD010_binary to numeric (1 = poor, 0 = good)
y = nhanes_scaled["HSD010_binary"].apply(lambda x: 1 if x == "poor" else 0)

# Define features by dropping target and non-feature columns
X = nhanes_scaled.drop(["HSD010_binary", "Cluster"], axis=1, errors="ignore")

# Partition data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)

```

```python

# Load libraries
import keras
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, Input

# Define the number of classes (binary classification: good vs. poor)from keras.models import Sequential
num_classes = 2
from keras.layers import Dense, Dropout

# Create neural network model
model = Sequential()

# Input layer
model.add(Input(shape=(X_train.shape[1],)))
model.add(Dense(64, input_dim=X_train.shape[1], kernel_initializer="he_normal", activation="relu"))

# Hidden layer
model.add(Dense(32, kernel_initializer="he_normal", activation="relu"))

# Dropout layer (optional)
model.add(Dropout(0.3))  # Adds regularization to prevent overfitting

# Output layer
model.add(Dense(num_classes, kernel_initializer="he_normal", activation="softmax"))

# Compile the model
from keras.utils import to_categorical

# Convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Confirm the number of classes
num_classes = y_test.shape[1]

# Print confirmation
print(f"Number of classes: {num_classes}")
```

```python
# create neural network model
# -----------
model = Sequential()

#input layer
model.add(Dense(64, input_dim=X_train.shape[1], kernel_initializer="he_normal", activation="relu"))

#hidden layer
model.add(Dense(32, kernel_initializer="he_normal", activation="relu"))

#additional dropout layer
model.add(Dropout(0.3))

#output layer
model.add(Dense(num_classes, kernel_initializer="he_normal", activation="softmax"))

model.compile(
    loss="categorical_crossentropy",  # Suitable for multi-class classification
    optimizer="adam",  # Adaptive learning rate optimizer
    metrics=["accuracy"]  # Track accuracy during training
)
history = model.fit(
    X_train, 
    y_train, 
    validation_data=(X_test, y_test), 
    epochs=20,  # Number of epochs
    batch_size=16,  # Batch size
    verbose=1  # Show training progress
)
```

### Neural Network on Principal Components

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=2)  # Adjust number of components as necessary
X_pca = pca.fit_transform(X)

# Partition PCA data
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.25, random_state=10)

# Convert y to one-hot encoding
y_train_pca = to_categorical(y_train_pca, num_classes)
y_test_pca = to_categorical(y_test_pca, num_classes)

# Create neural network model for PCA features
model_pca = Sequential()

# Input layer
model_pca.add(Input(shape=(X_train_pca.shape[1],)))
model_pca.add(Dense(64, kernel_initializer="he_normal", activation="relu"))

# Hidden layer
model_pca.add(Dense(32, kernel_initializer="he_normal", activation="relu"))

# Dropout layer
model_pca.add(Dropout(0.3))

# Output layer
model_pca.add(Dense(num_classes, kernel_initializer="he_normal", activation="softmax"))

# Compile the model
model_pca.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

# Train the model
history_pca = model_pca.fit(
    X_train_pca,
    y_train_pca,
    validation_data=(X_test_pca, y_test_pca),
    epochs=20,
    batch_size=16,
    verbose=1
)
```

### Neural Network on Principal Components + Cluster Membership

```python
from sklearn.cluster import KMeans

# Train KMeans on the original data
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Add clustering labels to PCA data
X_pca_clusters = np.hstack((X_pca, clusters.reshape(-1, 1)))

# Partition data
X_train_pca_clust, X_test_pca_clust, y_train_pca_clust, y_test_pca_clust = train_test_split(
    X_pca_clusters, y, test_size=0.25, random_state=10
)

# Convert y to one-hot encoding
y_train_pca_clust = to_categorical(y_train_pca_clust, num_classes)
y_test_pca_clust = to_categorical(y_test_pca_clust, num_classes)

# Create neural network model for PCA + Clustering features
model_pca_clust = Sequential()

# Input layer
model_pca_clust.add(Input(shape=(X_train_pca_clust.shape[1],)))
model_pca_clust.add(Dense(64, kernel_initializer="he_normal", activation="relu"))

# Hidden layer
model_pca_clust.add(Dense(32, kernel_initializer="he_normal", activation="relu"))

# Dropout layer
model_pca_clust.add(Dropout(0.3))

# Output layer
model_pca_clust.add(Dense(num_classes, kernel_initializer="he_normal", activation="softmax"))

# Compile the model
model_pca_clust.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

# Train the model
history_pca_clust = model_pca_clust.fit(
    X_train_pca_clust,
    y_train_pca_clust,
    validation_data=(X_test_pca_clust, y_test_pca_clust),
    epochs=20,
    batch_size=16,
    verbose=1
)
```

```python
# Comparing the results:

# Evaluate the original features model
test_loss_orig, test_acc_orig = model.evaluate(X_test, y_test)

# Evaluate the PCA model
test_loss_pca, test_acc_pca = model_pca.evaluate(X_test_pca, y_test_pca)

# Evaluate the PCA + Clustering model
test_loss_pca_clust, test_acc_pca_clust = model_pca_clust.evaluate(X_test_pca_clust, y_test_pca_clust)

# Print results
print(f"Original Features Model - Accuracy: {test_acc_orig:.2f}")
print(f"PCA Features Model - Accuracy: {test_acc_pca:.2f}")
print(f"PCA + Clustering Features Model - Accuracy: {test_acc_pca_clust:.2f}")
```

1. In your own words, what is the difference between PCA and clustering?

Essentially, principal component analysis is a tool that reduces the number of features by creating "principal components" that summarize the patterns in the original data. By doing so, this method keeps the focus on important information andd makes datasets easir to work with. Clustering on the other hand, just groups similar data points together by shared patterns. Clustring is useful for finding hidden structures in data. PCA helps simplify and summarize data while clustring helps find natural groupings within the dataset.
